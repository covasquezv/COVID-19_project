```python
from os import listdir
from os.path import isfile, join
import cv2
import matplotlib.pylab as plt
import numpy as np
```

##  Data

### PNEUMONIA - Kaggle

```python
pneum_path = '/home/cota/Documents/COVID-19/images/kaggle/chest_xray/PNEUMONIA/'
files_pneum = [f for f in listdir(pneum_path) if isfile(join(pneum_path, f))]
print(('# images (jpeg): {}').format(len(files_pneum)))
```

```python
x_, y_ = [], []

for file in files_pneum:
    im = cv2.imread(pneum_path+file)
    
    x_.append(im.shape[0])
    y_.append(im.shape[1])
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x_, y_)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Pneumonia (Kaggle)')
plt.show()
```

### COVID-19

```python
covid_path = '/home/cota/Documents/COVID-19/images/covid/original-ds/RX/'
files_covid = [f for f in listdir(covid_path) if isfile(join(covid_path, f))]

jpg, png, jpeg = 0, 0, 0

x_png, y_png = [], []
x_jpg, y_jpg = [], []
x_jpeg, y_jpeg = [], []

for file in files_covid:
    im = cv2.imread(covid_path+file)
    
    if 'jpeg' in file: 
        jpeg += 1
        x_jpeg.append(im.shape[0])
        y_jpeg.append(im.shape[1])
    
    elif 'jpg' in file: 
        jpg += 1
        x_jpg.append(im.shape[0])
        y_jpg.append(im.shape[1])
        
    elif 'png' in file: 
        png += 1
        x_png.append(im.shape[0])
        y_png.append(im.shape[1])

x_covid = x_png + x_jpg + x_jpeg
y_covid = y_png + y_jpg + y_jpeg
    
print(('# images (png): {}').format(png))
print(('# images (jpg): {}').format(jpg))
print(('# images (jpeg): {}').format(jpeg))
print(('Total: {}').format(len(files_covid)))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x_png, y_png, marker='s', alpha=.5)
plt.scatter(x_jpg, y_jpg, marker='.', alpha=.5)
plt.scatter(x_jpeg, y_jpeg, marker='*', alpha=.5)
plt.legend(('png', 'jpg', 'jpeg'), fontsize=16)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.title('COVID-19')
plt.show()
```
